```python
import time
import matplotlib.pyplot as plt
from codigo_original import encontrar_primos
from codigo_optimizado import encontrar_primos_optimizado

# Medir original
inicio = time.time()
primos_orig = encontrar_primos(1, 100000)
tiempo_orig = time.time() - inicio

# Medir optimizado
inicio = time.time()
primos_opt = encontrar_primos_optimizado(100000)
tiempo_opt = time.time() - inicio

print(f"Original: {tiempo_orig:.2f}s")
print(f"Optimizado: {tiempo_opt:.2f}s")
print(f"Mejora: {((tiempo_orig-tiempo_opt)/tiempo_orig*100):.1f}%")

# Gráfica 1: Barras comparativas
plt.figure(figsize=(8, 3))

plt.subplot(1, 3, 1)
nombres = ['Original', 'Optimizado']
tiempos = [tiempo_orig, tiempo_opt]
colores = ['#FF9999', '#99CCFF']
plt.bar(nombres, tiempos, color=colores)
plt.ylabel('Segundos')
plt.title('Comparación de Tiempos')
for i, v in enumerate(tiempos):
    plt.text(i, v + 0.1, f'{v:.2f}s', ha='center')

```

```python
# Gráfica 2: Gráfico de pastel
plt.subplot(1, 3, 3)
plt.pie(tiempos, labels=nombres, colors=colores, autopct='%1.1f%%')
plt.title('Distribución de Tiempos')

plt.tight_layout()
plt.savefig('graficas_comparacion.png')
plt.show()
```
